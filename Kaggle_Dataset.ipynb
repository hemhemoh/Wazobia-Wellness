```python
import torch
import pandas as pd
from sklearn.utils import shuffle
from transformers import MBartForConditionalGeneration, MBartTokenizer

# Assuming you have the model and tokenizer loaded already
# 'model' and 'tokenizer' should be defined somewhere else in your code
model.to(device)
# 'ibo_Latn','yor_Latn',
target_lang = ['hau_Latn']
batch_size = 15  # Adjust batch size as needed

def split_text_into_chunks(text, max_length):
    """Split a long text into smaller chunks with a maximum length"""
    tokens = tokenizer.encode(text)
    return [tokens[i:i + max_length] for i in range(0, len(tokens), max_length)]

for lang in target_lang:
    print(f"Processing translations for {lang}...")
    translated_data = []  # Store translations for the current language

    # Calculate the total number of batches
    total_batches = (len(data_json) + batch_size - 1) // batch_size
    for batch_number, start_idx in enumerate(range(0, len(data_json), batch_size), start=1):
        end_idx = min(start_idx + batch_size, len(data_json))
        batch_df = data_json.iloc[start_idx:end_idx]
        print(f"Processing batch {batch_number}/{total_batches}")

        # Tokenize the captions in the current batch and move to GPU
        contexts = batch_df['Context'].tolist()
        responses = batch_df['Response'].tolist()
        model_contexts = tokenizer(contexts, return_tensors="pt", padding=True, truncation=True).to(device)
        model_responses = tokenizer(responses, return_tensors="pt", padding=True, truncation=True).to(device)

        # Split contexts and responses into chunks if needed
        max_length = 1024  # Set max length as per model's token limit

        translated_contexts = []
        translated_responses = []

        # Translate contexts
        for context in contexts:
            context_chunks = split_text_into_chunks(context, max_length)
            context_translations = []
            for chunk in context_chunks:
                context_input = torch.tensor([chunk]).to(device)
                translated_chunk = model.generate(
                    context_input,
                    forced_bos_token_id=tokenizer.convert_tokens_to_ids(lang),
                    max_length=max_length
                )
                translated_chunk_text = tokenizer.decode(translated_chunk[0], skip_special_tokens=True)
                context_translations.append(translated_chunk_text)
            translated_contexts.append(" ".join(context_translations))  # Combine all chunk translations

        # Translate responses
        for response in responses:
            response_chunks = split_text_into_chunks(response, max_length)
            response_translations = []
            for chunk in response_chunks:
                response_input = torch.tensor([chunk]).to(device)
                translated_chunk = model.generate(
                    response_input,
                    forced_bos_token_id=tokenizer.convert_tokens_to_ids(lang),
                    max_length=max_length
                )
                translated_chunk_text = tokenizer.decode(translated_chunk[0], skip_special_tokens=True)
                response_translations.append(translated_chunk_text)
            translated_responses.append(" ".join(response_translations))  # Combine all chunk translations

        # Append the results to the list
        for i, _ in enumerate(translated_contexts):
            translated_data.append({
                "contexts": contexts[i],
                "responses": responses[i],
                "translated_contexts": translated_contexts[i],
                "translated_responses": translated_responses[i],
                "lang": lang
            })

    # Convert the list of dictionaries to a DataFrame for the current language
    df_translated = pd.DataFrame(translated_data)
    df_translated = shuffle(df_translated)

    # Save the DataFrame to a CSV file for the current language
    df_translated.to_csv(f'wazobia_dataset_{lang}.csv', index=False)
    print(f"Completed and saved translations for {lang} to wazobia_dataset_{lang}.csv")

print("Completed translations for all languages.")
```
